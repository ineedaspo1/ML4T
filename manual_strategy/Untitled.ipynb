```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime as dt
from util import get_data, plot_data
```

```python
sd = dt.datetime(2008,1,1)
ed = dt.datetime(2009,1,1)
syms = ['JPM']
sv=100000
#, rfr=0.0, sf=252.0):

# Read in adjusted closing prices for given symbols, date range
dates = pd.date_range(sd, ed)
prices_all = get_data(syms, dates)  # automatically adds SPY
price = prices_all[syms]            # only portfolio symbols   
```

```python
 #calc bb value
window_size = 20
k = 1
sma = price.rolling(window_size).mean()
std_bb = price.rolling(window_size).std()
bb = (price - sma) / (2*std_bb)
upper_b = bb + k*std_bb
lower_b = bb - k*std_bb
```

```python
df_bb = pd.concat([bb, upper_b, lower_b],\
                        keys = ['BB', 'Upper_BB', 'Lower_BB'],\
                        axis = 1)
plot_data(df_bb)
```

```python
# calc momentum
delta = 5
momentum = price/price.shift(delta) - 1
    
```
