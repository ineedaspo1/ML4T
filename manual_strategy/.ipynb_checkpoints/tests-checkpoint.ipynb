## Bollinger Bands

```python
 #calc bb value
window_size = 20
k = 1
sma = price.rolling(window_size).mean()
std_bb = price.rolling(window_size).std()
bb = (price - sma) / (2*std_bb)
upper_b = sma + k*std_bb
lower_b = sma - k*std_bb
```

```python
df_bb = pd.concat([price, sma, upper_b, lower_b],\
                        keys = ['Price', 'SMA', 'Upper_BB', 'Lower_BB'],\
                        axis = 1)
df_bb.dropna()
plot_data(df_bb)
```

```python
# calc momentum
delta = 5
momentum = price/price.shift(delta) - 1
std_m = momentum.std()    
```

```python
df_m = pd.concat([momentum, momentum + std_m, momentum - std_m],\
                        keys = ['Momentum', 'Upper_M', 'Lower_M'],\
                        axis = 1)
plot_data(df_m)
```
